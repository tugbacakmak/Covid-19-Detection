```python
!pip install -U scikit-fuzzy

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib
```

```python
# fuzzification of the input
# converting symptom input values into fuzzy sets
convulsion = ctrl.Antecedent(np.arange(0, 11, 1), 'convulsion')
distance = ctrl.Antecedent(np.arange(0, 10, 1), 'distance')
temperature = ctrl.Antecedent(np.arange(36, 40, 1), 'temperature')
chills = ctrl.Antecedent(np.arange(0,4,1), 'chills')
cough = ctrl.Antecedent(np.arange(0, 5, 1), 'cough')
fatigue = ctrl.Antecedent(np.arange(0, 10, 1), 'fatigue')
shortness_of_breath = ctrl.Antecedent(np.arange(0, 11, 1), 'shortness of breath')
loss_of_taste = ctrl.Antecedent(np.arange(0, 11, 1), 'loss of taste')
loss_of_smell = ctrl.Antecedent(np.arange(0, 11, 1), 'loss of smell')
muscle_weakness = ctrl.Antecedent(np.arange(0,4,1), 'muscle weakness')
sore_throat = ctrl.Antecedent(np.arange(0, 11, 1), 'sore throat')
headache = ctrl.Antecedent(np.arange(0, 11, 1), 'headache')
confusion = ctrl.Antecedent(np.arange(0,4,1), 'confusion')
nasal_congestion = ctrl.Antecedent(np.arange(0, 11, 1), 'nasal congestion')
diarrhea = ctrl.Antecedent(np.arange(0, 11, 1), 'diarrhea')
nausea = ctrl.Antecedent(np.arange(0,4,1), 'nausea')
sweating = ctrl.Antecedent(np.arange(0,4,1), 'sweating')
lack_of_appetite = ctrl.Antecedent(np.arange(0, 11, 1), 'lack of appetite')
dysphagia = ctrl.Antecedent(np.arange(0, 11, 1), 'dysphagia')
numbness = ctrl.Antecedent(np.arange(0, 11, 1), 'numbness')
vertigo = ctrl.Antecedent(np.arange(0,4,1), 'vertigo')
loss_of_hearing = ctrl.Antecedent(np.arange(0,4,1), 'loss of hearing')

# assing an entire fuzzy set to the output

isCovid = ctrl.Consequent(np.arange(0,10,1), 'isCovid')

# input using the membership funciton to define fuzzy sets in a fuzzy logic system
temperature['too low'] = fuzz.trimf(temperature.universe, [36, 36, 36.5])
temperature['low'] = fuzz.trimf(temperature.universe, [36.2, 36.5, 36.8])
temperature['optimum'] = fuzz.trimf(temperature.universe, [36.5, 36.8, 37.5])
temperature['high'] = fuzz.trimf(temperature.universe, [37.5, 38, 39])
temperature['too high'] = fuzz.trimf(temperature.universe, [38.5, 40 ,40])

chills['none']= fuzz.trimf(chills.universe, [0, 0, 0])
chills['mild'] = fuzz.trimf(chills.universe, [0, 0, 1])
chills['moderate'] = fuzz.trimf(chills.universe, [1, 2, 3])
chills['severe'] = fuzz.trimf(chills.universe, [3, 4, 5])

cough['none']= fuzz.trimf(cough.universe, [0, 0, 0])
cough['low']= fuzz.trimf(cough.universe, [0, 0, 1])
cough['medium'] = fuzz.trimf(cough.universe, [1, 2, 3])
cough['high']= fuzz.trimf(cough.universe, [3, 4, 5])

fatigue['none']= fuzz.trimf(fatigue.universe, [0, 0, 0])
fatigue['low'] = fuzz.trimf(fatigue.universe,[0, 0, 1])
fatigue['medium'] = fuzz.trimf(fatigue.universe,[1, 5, 6])
fatigue['high'] = fuzz.trimf(fatigue.universe,[6, 10, 11])

shortness_of_breath['none']= fuzz.trimf(shortness_of_breath.universe, [0, 0, 2])
shortness_of_breath['low'] = fuzz.trimf(shortness_of_breath.universe, [1, 2, 5])
shortness_of_breath['medium'] = fuzz.trimf(shortness_of_breath.universe, [2, 6, 8])
shortness_of_breath['high'] = fuzz.trimf(shortness_of_breath.universe, [7, 10, 10])

loss_of_taste['none']= fuzz.trimf(loss_of_taste.universe, [0, 0, 2])
loss_of_taste['mild'] = fuzz.trimf(loss_of_taste.universe, [1, 2, 5])
loss_of_taste['moderate'] = fuzz.trimf(loss_of_taste.universe, [2, 6, 8])
loss_of_taste['severe'] = fuzz.trimf(loss_of_taste.universe, [7, 10, 10])

loss_of_smell['none']= fuzz.trimf(loss_of_smell.universe, [0, 0, 2])
loss_of_smell['mild'] = fuzz.trimf(loss_of_smell.universe, [1, 2, 5])
loss_of_smell['moderate'] = fuzz.trimf(loss_of_smell.universe, [2, 6, 8])
loss_of_smell['severe'] = fuzz.trimf(loss_of_smell.universe, [7, 10, 10])

muscle_weakness['none']= fuzz.trimf(muscle_weakness.universe, [0, 0, 0])
muscle_weakness['mild'] = fuzz.trimf(muscle_weakness.universe, [0, 0, 1])
muscle_weakness['moderate'] = fuzz.trimf(muscle_weakness.universe, [1, 2, 3])
muscle_weakness['severe'] = fuzz.trimf(muscle_weakness.universe, [3, 4, 5])

sore_throat['none']= fuzz.trimf(sore_throat.universe, [0, 0, 2])
sore_throat['mild'] = fuzz.trimf(sore_throat.universe, [1, 2, 5])
sore_throat['moderate'] = fuzz.trimf(sore_throat.universe, [2, 6, 8])
sore_throat['severe'] = fuzz.trimf(sore_throat.universe, [7, 10, 10])

headache['none']= fuzz.trimf(headache.universe, [0, 0, 2])
headache['mild'] = fuzz.trimf(headache.universe, [1, 2, 5])
headache['moderate'] = fuzz.trimf(headache.universe, [2, 6, 8])
headache['severe'] = fuzz.trimf(headache.universe, [7, 10, 10])

confusion['none']= fuzz.trimf(confusion.universe, [0, 0, 0])
confusion['mild'] = fuzz.trimf(confusion.universe, [0, 1, 2])
confusion['moderate'] = fuzz.trimf(confusion.universe, [2, 3, 4])
confusion['severe'] = fuzz.trimf(confusion.universe, [4, 4, 5])

nasal_congestion['none']= fuzz.trimf(nasal_congestion.universe, [0, 0, 2])
nasal_congestion['mild'] = fuzz.trimf(nasal_congestion.universe, [1, 2, 5])
nasal_congestion['moderate'] = fuzz.trimf(nasal_congestion.universe, [2, 6, 8])
nasal_congestion['severe'] = fuzz.trimf(nasal_congestion.universe, [7, 10, 10])

diarrhea['none']= fuzz.trimf(diarrhea.universe, [0, 0, 2])
diarrhea['mild'] = fuzz.trimf(diarrhea.universe, [1, 2, 5])
diarrhea['moderate'] = fuzz.trimf(diarrhea.universe, [2, 6, 8])
diarrhea['severe'] = fuzz.trimf(diarrhea.universe, [7, 10, 10])

nausea['none']= fuzz.trimf(nausea.universe, [0, 0, 0])
nausea['mild'] = fuzz.trimf(nausea.universe, [0, 1, 2])
nausea['moderate'] = fuzz.trimf(nausea.universe, [2, 3, 4])
nausea['severe'] = fuzz.trimf(nausea.universe, [4, 4, 5])

sweating['none']= fuzz.trimf(sweating.universe, [0, 0, 0])
sweating['mild'] = fuzz.trimf(sweating.universe, [0, 1, 2])
sweating['moderate'] = fuzz.trimf(sweating.universe, [2, 3, 4])
sweating['severe'] = fuzz.trimf(sweating.universe, [4, 4, 5])

lack_of_appetite['none']= fuzz.trimf(lack_of_appetite.universe, [0, 0, 2])
lack_of_appetite['mild'] = fuzz.trimf(lack_of_appetite.universe, [1, 2, 5])
lack_of_appetite['moderate'] = fuzz.trimf(lack_of_appetite.universe, [2, 6, 8])
lack_of_appetite['severe'] = fuzz.trimf(lack_of_appetite.universe, [7, 10, 10])

dysphagia['none']= fuzz.trimf(dysphagia.universe, [0, 0, 2])
dysphagia['mild'] = fuzz.trimf(dysphagia.universe, [1, 2, 5])
dysphagia['moderate'] = fuzz.trimf(dysphagia.universe, [2, 6, 8])
dysphagia['severe'] = fuzz.trimf(dysphagia.universe, [7, 10, 10])

numbness['none']= fuzz.trimf(numbness.universe, [0, 0, 2])
numbness['mild'] = fuzz.trimf(numbness.universe, [1, 2, 5])
numbness['moderate'] = fuzz.trimf(numbness.universe, [2, 6, 8])
numbness['severe'] = fuzz.trimf(numbness.universe, [7, 10, 10])

vertigo['none']= fuzz.trimf(vertigo.universe, [0, 0, 0])
vertigo['mild'] = fuzz.trimf(vertigo.universe, [0, 1, 2])
vertigo['moderate'] = fuzz.trimf(vertigo.universe, [2, 3, 4])
vertigo['severe'] = fuzz.trimf(vertigo.universe, [4, 4, 5])

loss_of_hearing['none']= fuzz.trimf(loss_of_hearing.universe, [0, 0, 0])
loss_of_hearing['mild'] = fuzz.trimf(loss_of_hearing.universe, [0, 1, 2])
loss_of_hearing['moderate'] = fuzz.trimf(loss_of_hearing.universe, [2, 3, 4])
loss_of_hearing['severe'] = fuzz.trimf(loss_of_hearing.universe, [4, 4, 5])

# automatic membership function
convulsion.automf(3)
distance.automf(3)
isCovid.automf(3)

isCovid['unlikely'] = fuzz.trimf(isCovid.universe, [0, 0, 5])
isCovid['uncertain'] = fuzz.trimf(isCovid.universe, [2, 5, 8])
isCovid['likely'] = fuzz.trimf(isCovid.universe, [5, 10 ,10])

```

```python
temperature.view()
convulsion.view()
distance.view()
cough.view()
fatigue.view()
shortness_of_breath.view()
loss_of_taste.view()
loss_of_smell.view()
chills.view()
muscle_weakness.view()
sore_throat.view()
headache.view()
confusion.view()
nasal_congestion.view()
diarrhea.view()
nausea.view()
sweating.view()
lack_of_appetite.view()
dysphagia.view()
numbness.view()
vertigo.view()
loss_of_hearing.view()
isCovid.view()
```

```python
# fuzzy rules using the symptoms to determine whether patient is Covid-19 or not
rule1 = ctrl.Rule(temperature['high'] & distance['poor'], isCovid['likely'])
rule2 = ctrl.Rule(convulsion['average'] & headache['moderate'], isCovid['uncertain'])
rule3 = ctrl.Rule(temperature['high'] & convulsion['average'] & shortness_of_breath['high'], isCovid['likely'])
rule4 = ctrl.Rule(temperature['high'] & muscle_weakness['moderate'] & loss_of_taste['severe'],isCovid['likely'])
rule5 = ctrl.Rule(nausea['mild'] & headache['mild'],isCovid['poor'])

rule6 = ctrl.Rule(lack_of_appetite['severe'] & vertigo['moderate'] & temperature['high'],  isCovid['likely'])
rule7 = ctrl.Rule(cough['low'] & temperature['optimum'] & sore_throat['mild'],isCovid['poor'])
rule8 = ctrl.Rule(temperature['too high'] & confusion['moderate'] & nausea['moderate'], isCovid['likely'] )
rule9 = ctrl.Rule(loss_of_hearing['mild'] & lack_of_appetite['severe'], isCovid['uncertain'])
rule10 = ctrl.Rule(dysphagia['severe'] & muscle_weakness['severe'] & temperature['high'] , isCovid['likely'])

rule11 = ctrl.Rule(chills['moderate'] & vertigo['moderate'] & loss_of_smell['mild'], isCovid['uncertain'])
rule12 = ctrl.Rule(sore_throat['moderate'] & diarrhea['mild'] & loss_of_hearing['mild'], isCovid['uncertain'])
rule13 = ctrl.Rule(lack_of_appetite['moderate'] & loss_of_taste['severe'] & temperature['high'] , isCovid['likely'])
rule14 = ctrl.Rule(numbness['mild'] & vertigo['severe'], isCovid['unlikely'])
rule15 = ctrl.Rule(temperature['optimum'] & nasal_congestion['mild'] , isCovid['unlikely'])

rule16 = ctrl.Rule(chills['moderate'] & numbness['mild'], isCovid['unlikely'])
rule17 = ctrl.Rule(sore_throat['severe'] & headache['moderate'] & temperature['high'], isCovid['likely'])
rule18 = ctrl.Rule(nausea['moderate'] & vertigo['moderate'] & temperature['too high'], isCovid['likely'])
rule19 = ctrl.Rule(lack_of_appetite['moderate'] & loss_of_taste['none'], isCovid['uncertain'])
rule20 = ctrl.Rule(dysphagia['mild'] & sweating['moderate'] & temperature['optimum'] , isCovid['unlikely'])

rule21 = ctrl.Rule(fatigue['medium'] & loss_of_taste['moderate'] & muscle_weakness['moderate'],isCovid['likely'])
rule22 = ctrl.Rule(numbness['mild'] & sweating['none'] & sore_throat['none'], isCovid['unlikely'])
rule23 = ctrl.Rule(nausea['severe'] & vertigo['moderate'] & lack_of_appetite['moderate'], isCovid['uncertain'])
rule24 = ctrl.Rule(chills['mild'] & cough['low'], isCovid['unlikely'])
rule25 = ctrl.Rule(headache['moderate'] & confusion['moderate'] & muscle_weakness['severe'], isCovid['likely'])

# 2 & after use ',' otherwise you got error ??
```

```python

covid_ctrl = ctrl.ControlSystem([rule25,rule24,rule23,rule22,rule21,
                                 rule20,rule19,rule18,rule17,rule16,
                                 rule15,rule14,rule13,rule12,rule11,
                                 rule10, rule9, rule8, rule7, rule6,
                                 rule5, rule4,rule3, rule2, rule1  ])
```

```python
covid = ctrl.ControlSystemSimulation(covid_ctrl)
```
